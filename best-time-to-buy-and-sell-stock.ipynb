# 121. Best Time to Buy and Sell Stock

🔗 [LeetCode Link](https://leetcode.com/problems/best-time-to-buy-and-sell-stock/)  
**Difficulty:** 🟢 Easy  
**Topics:** Array, Dynamic Programming

---
## Problem Statement

You are given an array `prices` where `prices[i]` is the price of a given stock on the $i^{th}$ day.

You want to maximize your profit by choosing a **single day** to buy one stock and choosing a **different day in the future** to sell that stock.

Return *the maximum profit you can achieve from this transaction*. If you cannot achieve any profit, return `0`.

---
## Examples

### Example 1
```
Input: prices = [7, 1, 5, 3, 6, 4]
Output: 5
```
**Explanation:** Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5. Note that buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.

### Example 2
```
Input: prices = [7, 6, 4, 3, 1]
Output: 0
```
**Explanation:** In this case, no transactions are done and the max profit = 0.

---
## Constraints

- `1 <= prices.length <= 10⁵`
- `0 <= prices[i] <= 10⁴`

---
## Approaches to Consider

| # | Approach | Time | Space | Key Idea |
|---|----------|------|-------|----------|
| 1 | **Brute Force** | O(n²) | O(1) | Check every pair of buy and sell days |
| 2 | **One Pass / Greedy** | O(n) | O(1) | Keep track of the minimum price seen so far and the maximum profit |

---
## Key Intuition

Instead of comparing every possible pair of days, we can keep track of the **lowest price** we've seen so far. For every day, we calculate the profit we would make if we sold on that day (current price - lowest price seen so far). We then update the **maximum profit** if this profit is greater than our previous maximum.

```
prices = [7, 1, 5, 3, 6, 4]

Step 1: price=7 → min_price=7 → profit=0 → max_profit=0
Step 2: price=1 → min_price=1 → profit=0 → max_profit=0
Step 3: price=5 → min_price=1 → profit=4 → max_profit=4
Step 4: price=3 → min_price=1 → profit=2 → max_profit=4
Step 5: price=6 → min_price=1 → profit=5 → max_profit=5
Step 6: price=4 → min_price=1 → profit=3 → max_profit=5
```

---
## Edge Cases to Think About

- **Prices constantly decreasing:** `prices = [7, 6, 4, 3, 1]` — no profit can be made (return `0`)
- **Minimum length:** `prices = [1]` or `prices = [1, 2]` — array with 1 or 2 elements
- **Prices are the same:** `prices = [3, 3, 3]` — max profit is `0`

---
## Follow-up

> What if you could execute as many transactions as you like (LeetCode 122), or at most two transactions (LeetCode 123)?

---
## Solution

Write your solution below 👇

````python
class Solution:
    def maxProfit(self, prices):
        min_price = prices[0]
        max_profit = 0
        
        for price in prices:
            profit = price - min_price
            max_profit = max(max_profit, profit)
            min_price = min(min_price, price)
        
        return max_profit
        
````
