## Problem Statement

Given an array of integers `nums` and an integer `target`, return **indices of the two numbers** such that they add up to `target`.

You may assume that each input would have **exactly one solution**, and you may **not** use the same element twice.

You can return the answer in **any order**.

---
## Examples

### Example 1
```
Input:  nums = [2, 7, 11, 15], target = 9
Output: [0, 1]
```
**Explanation:** Because `nums[0] + nums[1] == 9`, we return `[0, 1]`.

### Example 2
```
Input:  nums = [3, 2, 4], target = 6
Output: [1, 2]
```

### Example 3
```
Input:  nums = [3, 3], target = 6
Output: [0, 1]
```

---
## Constraints

- `2 <= nums.length <= 10⁴`
- `-10⁹ <= nums[i] <= 10⁹`
- `-10⁹ <= target <= 10⁹`
- **Only one valid answer exists.**

---
## Approaches to Consider

| # | Approach | Time | Space | Key Idea |
|---|----------|------|-------|----------|
| 1 | **Brute Force** | O(n²) | O(1) | Check every pair of numbers |
| 2 | **Hash Map (One Pass)** | O(n) | O(n) | For each number, check if `target - num` is already in the map |
| 3 | **Hash Map (Two Pass)** | O(n) | O(n) | First pass builds the map, second pass looks up complements |
| 4 | **Sorting + Two Pointers** | O(n log n) | O(n) | Sort, then use two pointers (need to track original indices) |

---
## Key Intuition

Instead of searching for **two numbers that add up to target**, reframe it as:

> For each number `num`, does `target - num` (**the complement**) already exist in what we've seen?

This transforms a **search problem** into a **lookup problem**, which a hash map solves in O(1).

```
nums = [2, 7, 11, 15], target = 9

Step 1: num=2  → complement=7  → seen={}         → 7 not in seen → add {2: 0}
Step 2: num=7  → complement=2  → seen={2: 0}     → 2 IS in seen! → return [0, 1] ✅
```

- **Negative numbers:** `nums = [-1, -2, -3, -4, -5], target = -8`
- **Zero in array:** `nums = [0, 4, 3, 0], target = 0`
- **Minimum length:** `nums = [1, 2], target = 3` — only two elements
