# 15. 3Sum

🔗 [LeetCode Link](https://leetcode.com/problems/3sum/)  
**Difficulty:** 🟡 Medium  
**Topics:** Array, Two Pointers, Sorting

---
## Problem Statement

Given an integer array `nums`, return all the triplets `[nums[i], nums[j], nums[k]]` such that `i != j`, `i != k`, and `j != k`, and `nums[i] + nums[j] + nums[k] == 0`.

Notice that the solution set must **not contain duplicate triplets**.

---
## Examples

### Example 1
```
Input:  nums = [-1, 0, 1, 2, -1, -4]
Output: [[-1, -1, 2], [-1, 0, 1]]
```
**Explanation:** `nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0`, `nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0`, `nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0`. The distinct triplets are `[-1, 0, 1]` and `[-1, -1, 2]`.

### Example 2
```
Input:  nums = [0, 1, 1]
Output: []
```
**Explanation:** The only possible triplet does not sum up to `0`.

### Example 3
```
Input:  nums = [0, 0, 0]
Output: [[0, 0, 0]]
```
**Explanation:** The only possible triplet sums up to `0`.

---
## Constraints

- `3 <= nums.length <= 3000`
- `-10⁵ <= nums[i] <= 10⁵`

---
## Approaches to Consider

| # | Approach | Time | Space | Key Idea |
|---|----------|------|-------|----------|
| 1 | **Brute Force** | O(n³) | O(1) | Check every triplet combination, use a set to avoid duplicates |
| 2 | **Sorting + Two Pointers** | O(n²) | O(1) | Sort the array, fix one element, then use two pointers to find pairs that sum to its negation |
| 3 | **Hash Set** | O(n²) | O(n) | For each pair, check if the complement exists in a hash set |

---
## Key Intuition

The trick is to **reduce 3Sum to 2Sum** by fixing one number and solving 2Sum for the remaining target.

> Sort the array first. For each element `nums[i]`, use **two pointers** (`left = i+1`, `right = n-1`) to find pairs that sum to `-nums[i]`. Skip duplicates at every level to avoid duplicate triplets.

```
nums = [-1, 0, 1, 2, -1, -4]  →  sorted: [-4, -1, -1, 0, 1, 2]

Fix i=0 (nums[i]=-4): target=4, left=1, right=5 → no valid pair
Fix i=1 (nums[i]=-1): target=1, left=2, right=5
  → nums[2]+nums[5] = -1+2 = 1 ✅ → triplet [-1, -1, 2]
  → move pointers → nums[3]+nums[4] = 0+1 = 1 ✅ → triplet [-1, 0, 1]
Fix i=2 (nums[i]=-1): skip! same as nums[1]
```

---
## Edge Cases to Think About

- **All zeros:** `nums = [0, 0, 0]` — Only one triplet `[0, 0, 0]`
- **No valid triplet:** `nums = [1, 2, 3]` — Return `[]`
- **All same non-zero:** `nums = [1, 1, 1]` — Return `[]`
- **Duplicates requiring skip logic:** `nums = [-1, -1, -1, 2, 2]` — Only `[-1, -1, 2]`
- **Large negative and positive:** `nums = [-100000, 50000, 50000]` — Valid triplet

---
## Solution

**Sorting + Two Pointers** — O(n²) time, O(1) extra space ✅

````python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        result = []

        for i in range(len(nums) - 2):
            # Skip duplicates for the first element
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            # Early termination: if smallest value is positive, no triplet can sum to 0
            if nums[i] > 0:
                break

            left, right = i + 1, len(nums) - 1

            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total < 0:
                    left += 1
                elif total > 0:
                    right -= 1
                else:
                    result.append([nums[i], nums[left], nums[right]])
                    # Skip duplicates for second and third elements
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
                    left += 1
                    right -= 1

        return result
````

## Test Cases

````python
sol = Solution()

# Example test cases
result1 = sol.threeSum([-1, 0, 1, 2, -1, -4])
assert sorted([sorted(t) for t in result1]) == [[-1, -1, 2], [-1, 0, 1]], "Test 1 failed"

assert sol.threeSum([0, 1, 1]) == [], "Test 2 failed"
assert sol.threeSum([0, 0, 0]) == [[0, 0, 0]], "Test 3 failed"

# Edge cases
assert sol.threeSum([1, 2, 3]) == [], "Test 4 (no valid triplet) failed"

result5 = sol.threeSum([-1, -1, -1, 2, 2])
assert sorted([sorted(t) for t in result5]) == [[-1, -1, 2]], "Test 5 (duplicates) failed"

print("✅ All test cases passed!")
````
