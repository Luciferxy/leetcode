# 3. Longest Substring Without Repeating Characters

🔗 [LeetCode Link](https://leetcode.com/problems/longest-substring-without-repeating-characters/)  
**Difficulty:** 🟡 Medium  
**Topics:** Hash Table, String, Sliding Window

---
## Problem Statement

Given a string `s`, find the length of the **longest substring** without repeating characters.

---
## Examples

### Example 1
```
Input: s = "abcabcbb"
Output: 3
```
**Explanation:** The answer is `"abc"`, with the length of `3`.

### Example 2
```
Input: s = "bbbbb"
Output: 1
```
**Explanation:** The answer is `"b"`, with the length of `1`.

### Example 3
```
Input: s = "pwwkew"
Output: 3
```
**Explanation:** The answer is `"wke"`, with the length of `3`. Notice that `"pwke"` is a subsequence and not a substring.

---
## Constraints

- `0 <= s.length <= 5 × 10⁴`
- `s` consists of English letters, digits, symbols and spaces.


---
## Approaches to Consider

| # | Approach | Time | Space | Key Idea |
|---|----------|------|-------|----------|
| 1 | **Brute Force** | O(n³) | O(min(n, m)) | Check all substrings and verify each has no duplicates using a set |
| 2 | **Sliding Window + Set** | O(n) | O(min(n, m)) | Expand right pointer; when a duplicate is found, shrink from left until the duplicate is removed |
| 3 | **Sliding Window + HashMap** | O(n) | O(min(n, m)) | Store the last seen index of each character. When a duplicate is found, jump `left` directly past the previous occurrence |

---
## Key Intuition

For Sliding Window + HashMap:
Maintain a window `[left, right]` and a hashmap that maps each character to its **last seen index**.

> When we encounter a character that's already in the map and its index is `>= left`, we jump `left` to `last_seen[ch] + 1` — skipping over all characters up to and including the duplicate, instead of shrinking one by one.

At each step, update the result as `max(result, right - left + 1)`. This guarantees a single O(n) pass.

---
## Edge Cases to Think About

- **Empty string:** `s = ""` — Return `0`.
- **All same characters:** `s = "aaaa"` — Return `1`.
- **All unique characters:** `s = "abcdef"` — Return `len(s)`.
- **Spaces and symbols:** `s = " "` — A single space counts as a valid character.

---
## Solution

Write your solution below 👇

````python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        last_seen = {}
        left = 0
        result = 0

        for right, ch in enumerate(s):
            if ch in last_seen and last_seen[ch] >= left:
                left = last_seen[ch] + 1

            last_seen[ch] = right
            result = max(result, right - left + 1)

        return result
````
