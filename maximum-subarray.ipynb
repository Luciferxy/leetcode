## Problem Statement

Given an integer array `nums`, find the subarray with the largest sum, and return its sum.

---
## Examples

### Example 1
```
Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
Output: 6
```
**Explanation:** The subarray `[4,-1,2,1]` has the largest sum `6`.

### Example 2
```
Input: nums = [1]
Output: 1
```
**Explanation:** The subarray `[1]` has the largest sum `1`.

### Example 3
```
Input: nums = [5,4,-1,7,8]
Output: 23
```
**Explanation:** The subarray `[5,4,-1,7,8]` has the largest sum `23`.

---
## Constraints

- `1 <= nums.length <= 10⁵`
- `-10⁴ <= nums[i] <= 10⁴`


---
## Approaches to Consider

| # | Approach | Time | Space | Key Idea |
|---|----------|------|-------|----------|
| 1 | **Brute Force** | O(n²) | O(1) | Check every contiguous subarray sum |
| 2 | **Kadane's Algorithm** | O(n) | O(1) | Keep a running sum. If the sum drops below 0, reset it, as a negative prefix will always decrease the next element's value. Keep track of the maximum sum seen so far |
| 3 | **Divide and Conquer** | O(n log n) | O(log n) | Recursively find max in left half, right half, and crossing middle |

---
## Key Intuition

For Kadane's Algorithm:
Instead of looking at all subarrays, iterate through the array and maintain a `current_sum`.

> If the `current_sum` becomes negative, it's dragging down whatever comes next. So, we should abandon the previous subarray and start a new sequence from the current element.

At each step, take `current_sum = max(num, current_sum + num)` and update `max_sum = max(max_sum, current_sum)`.

---
## Edge Cases to Think About

- **All negative numbers:** `[-5, -2, -9]` — The algorithm should return `-2` instead of `0`. (Don't initialize `max_sum` to 0!)
- **Single element:** `[1]` or `[-1]` — Handled naturally but good to check.
