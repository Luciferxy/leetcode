# 76. Minimum Window Substring

🔗 [LeetCode Link](https://leetcode.com/problems/minimum-window-substring/)  
**Difficulty:** 🔴 Hard  
**Topics:** Hash Table, String, Sliding Window

---
## Problem Statement

Given two strings `s` and `t` of lengths `m` and `n` respectively, return the minimum window substring of `s` such that every character in `t` (including duplicates) is included in the window. If there is no such substring, return the empty string `""`.

The testcases will be generated such that the answer is **unique**.

---
## Examples

### Example 1
```
Input: s = "ADOBECODEBANC", t = "ABC"
Output: "BANC"
```
**Explanation:** The minimum window substring `"BANC"` includes `'A'`, `'B'`, and `'C'` from string `t`.

### Example 2
```
Input: s = "a", t = "a"
Output: "a"
```
**Explanation:** The entire string `s` is the minimum window.

### Example 3
```
Input: s = "a", t = "aa"
Output: ""
```
**Explanation:** Both `'a'`s from `t` must be included in the window. Since the largest window of `s` only has one `'a'`, return empty string.

---
## Constraints

- `m == s.length`
- `n == t.length`
- `1 <= m, n <= 10⁵`
- `s` and `t` consist of uppercase and lowercase English letters.


---
## Approaches to Consider

| # | Approach | Time | Space | Key Idea |
|---|----------|------|-------|----------|
| 1 | **Brute Force** | O(m² · n) | O(m + n) | Check every substring of `s` and verify if it contains all characters of `t` |
| 2 | **Sliding Window** | O(m + n) | O(1) | Use two pointers to expand and contract a window. Track character counts and shrink the window once all required characters are found |
| 3 | **Optimized Sliding Window** | O(m + n) | O(1) | Same as above but filter `s` to only include characters present in `t`, reducing unnecessary comparisons |

---
## Key Intuition

For Sliding Window:
Use a `need` counter for characters in `t` and a `window` counter for the current window in `s`. Expand the right pointer to include characters until all required characters are satisfied.

> Once the window contains all required characters (`formed == required`), **contract** from the left to find the smallest valid window. Track the minimum window seen so far.

The key insight is that `formed` counts the number of **unique** characters in `t` that are fully satisfied (i.e., their count in the window meets or exceeds their count in `t`). The window is valid when `formed == required`.

---
## Edge Cases to Think About

- **`t` longer than `s`:** `s = "a", t = "aa"` — Impossible to form a valid window, return `""`.
- **Exact match:** `s = "a", t = "a"` — The entire string is the answer.
- **Duplicate characters in `t`:** `t = "AABC"` — The window must contain at least two `'A'`s.

---
## Follow-up

> Could you find an algorithm that runs in `O(m + n)` time?

---
## Solution

Write your solution below 👇

````python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if not t or not s:
            return ""

        need = {}
        for ch in t:
            need[ch] = need.get(ch, 0) + 1
        required = len(need)

        left = formed = 0
        window = {}
        res = (float('inf'), 0, 0)

        for right, ch in enumerate(s):
            window[ch] = window.get(ch, 0) + 1
            if ch in need and window[ch] == need[ch]:
                formed += 1

            while formed == required:
                if right - left + 1 < res[0]:
                    res = (right - left + 1, left, right)

                left_ch = s[left]
                window[left_ch] -= 1
                if left_ch in need and window[left_ch] < need[left_ch]:
                    formed -= 1
                left += 1

        return "" if res[0] == float('inf') else s[res[1]:res[2] + 1]
````
