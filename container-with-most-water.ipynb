# Container With Most Water

LeetCode 11

This notebook explains the problem in a very simple way and shows the best solution.

## Problem in baby words

Think of each number as a wall.

If you pick **two walls**, water can sit between them.

How much water can they hold?

`water = width * shorter wall`

- `width` = distance between the two walls
- `shorter wall` = water spills over the shorter side, so that side limits everything

Your job: pick the best two walls so the water is as big as possible.

## Example

For:

`[1,8,6,2,5,4,8,3,7]`

If we pick `8` and `7`:

- width = `8 - 1 = 7`
- shorter wall = `7`
- water = `7 * 7 = 49`

So the answer is `49`.

## Smart idea: two fingers

Put one finger at the **left end** and one finger at the **right end**.

Why start there?

Because that gives the **biggest possible width**.

Then repeat:

1. Calculate water between the two walls.
2. Save the best answer.
3. Move the **shorter wall** inward.

Keep going until the two fingers meet.

## Why move the shorter wall?

Because the shorter wall is the bottleneck.

If left wall is `2` and right wall is `9`, water is limited by `2`.

If you move the tall wall:

- width becomes smaller
- short wall is still `2`

That cannot help.

So the only hope is:

move the shorter wall and try to find a taller one.

```python
def maxArea(height):
    left = 0
    right = len(height) - 1
    best = 0

    while left < right:
        width = right - left
        h = min(height[left], height[right])
        best = max(best, width * h)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return best


class Solution:
    def maxArea(self, height):
        return maxArea(height)
```

## Step-by-step walkthrough

Take:

`[1,8,6,2,5,4,8,3,7]`

Start:

- left at `1`
- right at `7`
- width = `8`
- area = `8 * 1 = 8`

The shorter wall is `1`, so move left.

Now:

- left at `8`
- right at `7`
- width = `7`
- shorter wall = `7`
- area = `7 * 7 = 49`

Best = `49`

Keep going, but nothing beats `49`.

```python
heights = [1, 8, 6, 2, 5, 4, 8, 3, 7]
print(Solution().maxArea(heights))
```

## Why this works

Each step throws away a bad choice safely.

If one side is shorter, keeping it while moving the taller side cannot create a better answer, because:

- width gets smaller
- the shorter wall still limits the water

So we always move the shorter pointer.

That lets us check all useful possibilities without trying every pair.

## Complexity

- Time: `O(n)` because each pointer moves at most `n` times
- Space: `O(1)` because we only use a few variables

## Tiny memory trick

**Wide first, then fix the short wall.**

That is the whole idea.
